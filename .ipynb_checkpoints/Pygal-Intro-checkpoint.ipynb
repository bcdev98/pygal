```python
import pygal
import os
import webbrowser
```

```python
bar_chart = pygal.Bar()
bar_chart.title = "Pet Population Per 100 Household"
bar_chart.add("Dogs", 41)
```

```python
bar_chart.add('Cats', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 2)
bar_chart.add('No Pets', 18)
```

```python
bar_chart.render_to_file('pet_bar_chart.svg')
```

```python
def open_in_browser(file_to_open):
    file_path = os.path.realpath(file_to_open)
    file_url = 'file://' + file_path
    browser = webbrowser.get('chrome')
    browser.open(file_url)
```

```python
open_in_browser('pet_bar_chart.svg')
```

```python
horizontal_bar = pygal.HorizontalBar(height=360, width=640, explict_size=True)
horizontal_bar.title = "Pet Population"
```

```python
horizontal_bar.add("Dogs", 41)
horizontal_bar.add("Birds", 12)
horizontal_bar.add("Horses", 7)
horizontal_bar.add("Exotic Pets", 2)
horizontal_bar.add("No Pets", 18)
```

```python
horizontal_bar.render_to_file("pet_horizontal_bar_chart.svg")
```

```python
open_in_browser('pet_horizontal_bar_chart.svg')
```

## Dataset Source
https://www.kaggle.com/ucim/red-wine-quality-cortez-et-all-2009 or search for winequality-red.csv

```python
import pygal
import pandas as pd
```

```python
# why use pd.read_csv(), explain
# tip - using tab to auto-fill 
wine_dataframe = pd.read_csv('./datasets/winequality-red.csv', sep=';')
wine_dataframe.head()
```

```python
wine_dataframe.describe()
```

```python
average_df = wine_dataframe.groupby(['quality']).mean()
```

```python
average_df
```

```python
bar_chart = pygal.Bar(height=360, width=640, explicit_size=True)
bar_chart.title = "Average Acidity by Wine Quality"
```

```python
bar_chart.x_title = 'Quality Rating'
bar_chart.x_labels = range(3, 9)
```

```python
bar_chart.add('Fixed Acidity', average_df['fixed acidity'])
```

```python
bar_chart.add('Volatity Acidity', average_df['volatile acidity'] )
# or add it using a list
# [0.88, 0.69, 0.58, 0.5, {'value':0.4}, 0.42]
```

```python
bar_chart = pygal.Bar(height=360, width=640, explicit_size=True, truncate_legend=-1)
```

```python
bar_chart.x_labels = range(3, 9)
bar_chart.title = "Average Acidity By Wine Quality"
bar_chart.x_title = 'Quality Rating'
```

```python
bar_chart.add('Fixed Acidity', average_df['fixed acidity'])
bar_chart.add('Volatile Acidity', average_df['volatile acidity'])
```

```python
bar_chart.render_in_browser()
```

```python
stack_bar = pygal.StackedBar(height=360, width=640, explict_size = True, truncate_legend=-1)
```

```python
stack_bar.x_labels = range(3, 9)
stack_bar.title = 'Average Acidity By Wine Quality'
stack_bar.x_title = "Quality Rating"
```

```python
stack_bar.add('Fixed Acidity', average_df['fixed acidity'])
stack_bar.add('Volatility Acidity', average_df['volatile acidity'])
```

```python
stack_bar.render_in_browser()
```
