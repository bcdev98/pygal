```python
import pygal
from IPython.display import display, HTML
```

```python
html_skeleton = """
<!DOCTYPE html>
<html>
    <head>
    <script type="text/javascript"
            src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js">
    </script>
    <script type="text/javascript"
            src="https://kozea.github.io/pygal.js/2.0.x/pygal-tooltips.min.js">
    </script>
    <head>
        <body>
            <figure>
                {rendered_chart}
            </figure>
        </body>
    </head>
</html>
"""

def display_chart(chart):
    rendered_chart = chart.render(is_unicode=True)
    plot_html = html_skeleton.format(rendered_chart=rendered_chart)
    display(HTML(plot_html))
```

```python
bar_chart = pygal.Bar(width=640, height=360, explicit_size=True)
```

```python
bar_chart.title = 'Pet Population by Animal Per 100 households'
```

```python
display_chart(bar_chart)
```

```python
bar_chart = pygal.Bar(width=640, height=360, explicit_size=True, no_data_text='Data Missing :(')
```

```python
display(bar_chart)
```

```python
bar_chart = pygal.Bar(width=640, height=360, explicit_size=True, no_data_text='Data Missing :(', print_values=True)
bar_chart.title = "Pet Population By Animal Per 100 households"
```

```python
bar_chart.add('Dogs', 41, rounded_bars=10)
display_chart(bar_chart)
```

```python
bar_chart.add('Cats', 20, rounded_bars=50, formatter=lambda x: '%s Cats' % x)
display_chart(bar_chart)
```

```python
# add 3nd bar using a dictionary object and spec out sytle 
dict_obj_list = [{'value':12, 'style': 'fill: green; stroke: black; stroke-width: 4'}]
```

```python
bar_chart.add('Birds', dict_obj_list)
display_chart(bar_chart)
```

```python
# get the horse list
horse_list = [{'value': 7, 'xlink': 'https://www.bluecross.org.uk/rehome/horse'}]
bar_chart.add('Horses', horse_list)
display_chart(bar_chart)
```

```python
# add pets
bar_chart.add('Exotic Pets', [{'value': 2, 'color': 'maroon'}])
display_chart(bar_chart)
```

```python

bar_chart.add({'title': 'No Pets', 'xlink': {'href': 'https://www.bluecross.org.uk/rehome-pet'}}, 18)
display_chart(bar_chart)
```

## style configuration

```python
from pygal.style import DarkStyle
```

```python
bar_chart = pygal.Bar(width=640, height = 360, explicit_size = True)
```

```python
bar_chart.style = DarkStyle
```

```python
bar_chart.add('Dog', 41)
bar_chart.add('Cat', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 2)
bar_chart.add('No Pets', 18)
display_chart(bar_chart)
```

```python
from pygal.style import DarkenStyle
```

```python
darken_style = DarkenStyle('#75FFFD')
```

```python
bar_chart = pygal.Bar(width=640, height=360, explicit_size=True, style=darken_style)
```

```python
bar_chart.add('Dogs', 41)
bar_chart.add('Cats', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 18)

display_chart(bar_chart)
```

```python
bar_chart.style = DarkenStyle('#75FFFD', step=5)
display_chart(bar_chart)
```

```python
from pygal.style import Style
```

```python
custom_style = Style(background='#ECF2F2',
                    plot_backround='transparent',
                    opacity='.6',
                    transition='400ms ease-in',
                    legend_font_family='arial',
                    legend_font_size=12,
                    colors=('red', '#8299F7', '#ED8223', 'purple', '#E809B53', 'deeppink'))
```

```python
bar_chart = pygal.Bar(width=640, height=360, explict_size=True, sytle=custom_style)
```

```python
bar_chart.add('Dogs', 41)
bar_chart.add('Cats', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 18)

display_chart(bar_chart)
```

```python
# set up a configuration
from pygal import Config
```

```python
custom_config = Config(width=640, height=360, explict_size=True, style=custom_style)
```

```python
bar_chart = pygal.Bar(config=custom_config)
```

```python
bar_chart.add('Dogs', 41)
bar_chart.add('Cats', 20)
bar_chart.add('Bird', 12)
bar_chart.add('Horses', 7)
bar_chart.add('Exotic Pets', 18)

display_chart(bar_chart)
```
